```python
import pandas as pd
import numpy as np

# Création des dates sur une période de 5 ans (52 semaines par an)
dates = pd.date_range(start='2018-01-01', end='2023-12-31', freq='W')
tables_max = 80
# Génération des données pour chaque semaine
donnees = {
    'Date': [],
    'Chiffre_daffaires': [],
    'Depenses': [],
    'Nb_Tables_Occ': [],
    'Nb_Tables_Tot': [],
    'Cout_MO': [],
    'VA': [],

}

for date in dates:
    depenses = {
        'Nourriture': np.random.randint(300, 1500),
        'Boissons': np.random.randint(500, 900),
        'Personnel': np.random.randint(600, 800),
        'Loyer': np.random.randint(200, 1000),
        'Autres': np.random.randint(100, 500)
    }
    total_depenses = sum(depenses.values())
    
    profit_rate = 0  # Initialiser le taux de profit
    tables_tot = 0
    tables_occ = 0
    
    # Déterminer le taux de profit en fonction du mois
    if date.month in [6, 7, 8]:  # Juin, Juillet, Août
        profit_rate = np.random.uniform(1.10, 1.35)  # Entre 80% et 120%
        tables_tot = tables_max
        tables_occ = tables_tot * np.random.uniform(0.80, 0.97)
        depenses['Personnel'] = depenses['Personnel']*1.2
    elif date.month in [9, 10, 3, 4, 5]:  # Septembre, Octobre, Mars, Avril, Mai
        profit_rate = np.random.uniform(1.05, 1.1)  # Entre 20% et 60%
        tables_tot = tables_max * np.random.uniform(0.65, 0.75)
        tables_occ = tables_tot * np.random.uniform(0.60, 0.75)
    else:
        profit_rate = np.random.uniform(1.02, 1.05)  # Entre -20% et 20%
        tables_tot = tables_max * np.random.uniform(0.55, 0.65)
        tables_occ = tables_tot * np.random.uniform(0.50, 0.60)
        depenses['Personnel'] = depenses['Personnel']*0.8


    total_depenses = sum(depenses.values())

    
    # Calculer le chiffre d'affaires ajusté en fonction du taux de profit
    if profit_rate >= 0:
        chiffre_daffaires_ajuste = total_depenses * profit_rate
    else:
        chiffre_daffaires_ajuste = total_depenses * (1-abs(profit_rate))

    donnees['Date'].append(date.strftime('%Y-%m-%d'))  # Format date without hours
    donnees['Chiffre_daffaires'].append(chiffre_daffaires_ajuste)
    donnees['Depenses'].append(total_depenses)
    donnees['Nb_Tables_Occ'].append(round(tables_occ))
    donnees['Nb_Tables_Tot'].append(round(tables_tot))
    donnees['Cout_MO'].append(depenses['Personnel'])
    donnees['VA'].append(depenses['Nourriture']+depenses['Boissons']+depenses['Loyer']+depenses['Autres'])


# Création d'un DataFrame à partir des données
df = pd.DataFrame(donnees)

# Calcul du taux de profit (marge ou coûts sur chiffre d'affaires)
df['Taux_de_profit'] = (df['Chiffre_daffaires'] - df['Depenses']) / df['Chiffre_daffaires']
df['ratio_MO_sur_VA'] = df['Cout_MO']/df['VA']
df['taux_occ_tables'] = df['Nb_Tables_Occ']/df['Nb_Tables_Tot']
df['CA_par_table'] = df['Chiffre_daffaires']/df['Nb_Tables_Occ']

# Nom du fichier Excel de sortie
fichier_excel = 'donnees_restaurant.xlsx'

# Écriture des données dans un fichier Excel
df.to_excel(fichier_excel, index=False)

print("Le fichier Excel a été généré avec succès : ", fichier_excel)
```

```python
import random
donnees = {
    'nom': [],
}
file_path = 'matiere_premirere_resto.xlsx'

df = pd.read_excel(file_path)

donnees['nom'] = df['nom'].tolist()

price = df['prix'].tolist()
dates = pd.date_range(start='2018-01-01', end='2023-12-31', freq='M')


y = 2018
for date in dates:
    inflation = (date.year - y)/100
    updated_prices = []
    if date.month in [6, 7, 8]:  # Juin, Juillet, Août
        for i in range (len(price)):
            updated_prices.append(price[i] * (random.uniform(1.05, 1.25))+inflation)
    elif date.month in [9, 10, 3, 4, 5]:  # Septembre, Octobre, Mars, Avril, Mai
        for i in range (len(price)):    
            updated_prices.append(price[i] * (random.uniform(1.01, 1.1))+inflation)

    else:
        for i in range (len(price)):
            updated_prices.append(price[i] * (random.uniform(0.9, 1.00))+inflation)

    date_str = pd.Timestamp(date).strftime('%Y-%m')
    donnees[str(date_str)] = updated_prices


f = pd.DataFrame(donnees)
frame = f.T

fichier_excel = 'donnees_restaurant_prix_produits.xlsx'

# Écriture des données dans un fichier Excel
frame.to_excel(fichier_excel, index=True)

print("Le fichier Excel a été généré avec succès : ", fichier_excel)
```
